```python
#Kenny Weeks
import matplotlib.pyplot as plt
import numpy as np

if __name__ == "__main__":
	N = 300 #This is the number of spaces in between stuff
	x = np.arange(-2, 1 + (3 / N), 3 / N)
	y = np.arange(-1, 1 + (2 / N), 2 / N)

	#Number of iterations to determine divergence or convergence
	it = 50

	#This is a list of magnitudes are the result 
	mag = list()

	#python abs() about a complex number gives it's magnitude
	for b in y:
		mag.append(list())
		for a in x:
			cy = b #Y on the plop
			cx = a#X on the plot
			zx = 0
			zy = 0
			n = 0 #Iteration counter
			while n < it:
				nx = (zx**2) - (zy**2) + cx 
				ny = (2 * zy * zx) + cy
				if (nx**2) + (ny**2) > 2:
					#plt.plot(a, b, 'bo')
					mag[-1].append(0) #add iteration here as a way of determining divergence
					break;
				zx = nx
				zy = ny
				n += 1

			if n == it:
				mag[-1].append(it) #zero is representative of convergence
				#plt.plot(a, b, "wx")

	#plt.axis([-2, 2, -2, 2])
	#plt.show()
	plt.imshow(mag, extent=[-2, 1, -1, 1], cmap="binary", interpolation="bilinear")
	plt.show()
	print("Done")
	

	#magnitude is just a^2 + b^s
```
