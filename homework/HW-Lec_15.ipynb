```python
import matplotlib.pyplot as plt
import math
import numpy as np
from tabulate import tabulate
```

```python
#%matplotlib notebook 
#Graphing code
x_min, x_max = -2, 2                          # range of x
y_min, y_max = -2, 2                          # range of x

def f2(x,y):
    return 1/2*x**2 + 1/4*y**2

nx = np.linspace(x_min,x_max,100)
ny = np.linspace(y_min,y_max,100)

x,y = np.meshgrid(nx,ny)

z = f2(x, y)
fig = plt.figure()
fig.set_figheight(5)
fig.set_figwidth(5)
ax = fig.gca(projection='3d')
ax.plot_surface(x,y,z, cmap='summer', alpha=0.8)
cset = ax.contour(x,y,z, zdir='z', offset=0)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('f')

plt.show()
```

```python
#this is the monte carlo approximation of the minimum
#The approximation allows for a simple way to find the minimum or maximum of a set of data with random number sampling


def MonteCarlo(x0, y0, x1, x2, y1, y2, xm, ym, m):
    lst = list()
    res = list()
    for i in range(10):
        x = x0 - np.random.random()*(x_min - x_max) #Create the point based on a percentage of the entire domain
        y = y0 - np.random.random()*(y_min - y_max) #Create the point based on a percentage of the entire range
        
        
        result = f2(x, y) #Get the result of the list
        res.append(result)
        ind = [x, y, result] #Save the entire explanation of the current graph and coordinates
        lst.append(ind) #Append this to the list
        
        m = min(res)
        xm = lst[res.index(m)][0]
        ym = lst[res.index(m)][1]
        
    return m, xm, ym, lst
                
        
m, xm, ym, lst = MonteCarlo(0, 0, x_min, x_max, y_min, y_max, 0, 0, -1)
print("Miniumum Coordinates --> " + str(xm) + ", " + str(ym))
print("Minimum Value --> " + str(m))

print(tabulate(lst, headers=["X", "Y", "f(X, Y)"], tablefmt="grid"))
    
    
```
