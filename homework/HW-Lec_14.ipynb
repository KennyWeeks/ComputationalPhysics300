```python
from matplotlib.animation import FuncAnimation
import matplotlib.pyplot as plt
import numpy as np
```

```python
N=10000
L1 = 101
L2 = 101

xdata = []
ydata = []
direction = [[0, 1], [0, -1], [1, 0], [-1, 0]]

xdata.append(int(L1/2))
ydata.append(int(L2/2))

for i in range(1, N):
    x = -1
    y = -1
    
    while min(x, y) < 0 or x > L1 or y > L2:
        move = np.random.randint(len(direction))
        x = xdata[i-1] + direction[move][0]
        y = ydata[i-1] + direction[move][1]
        
    xdata.append(x)
    ydata.append(y)

plt.figure(figsize=[5, 5])
plt.xlim(0, L1)
plt.ylim(0, L2)
plt.plot(xdata, ydata, "ro")
plt.show()
```

```python
#Animation Begins Here
fig, ax = plt.subplots()
N=10000
L1 = 101
L2 = 101

print(np.arange(1, N))

xdata = []
ydata = []
ln, = plt.plot([], [], "ro")
direction = [[0, 1], [0, -1], [1, 0], [-1, 0]]

def init():
    ax.set_xlim(0, L1)
    ax.set_ylim(0, L2)
    xdata.append(int(L1/2))
    ydata.append(int(L2/2))
    ln.set_data(xdata, ydata)
    return ln,


def update(frame):
    x = -1
    y = -1
    
    while min(x, y) < 0 or x > L1 or y > L2:
        move = np.random.randint(len(direction))
        x = xdata[frame-1] + direction[move][0]
        y = ydata[frame-1] + direction[move][1]
        
    xdata.append(x)
    ydata.append(y)
    ln.set_data(xdata, ydata)
    return ln,

ani = FuncAnimation(fig, update, frames=np.arange(1, N), init_func=init, blit=True)
plt.show()
```
