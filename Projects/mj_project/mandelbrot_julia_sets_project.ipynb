```python
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
def mandelBrot():
    N = 300 #This is the number of spaces in the range
    x = np.arange(-2, 2 + (3/N), 3/N) #This is the linspace of the x axis
    y = np.arange(-2, 2 + (3/N), 3/N) #This is the linspace of the y axis
    
    it = 50 #This is the number of interations
    
    mag = list()
    
    for b in y:
        mag.append(list()) #This is adding the row to the magnitude matrix
        for a in x:
            cy = b
            cx = a
            zx = 0
            zy = 0
            n = 0
            while n < it:
                nx = (zx**2) - (zy**2) + cx
                ny = (2 * zy * zx) + cy
                if (nx**2) + (ny**2) > 2:
                    mag[-1].append(0)
                    break
                zx = nx
                zy = ny
                n += 1
                
            if n == it:
                mag[-1].append(it)
                
    plt.imshow(mag, extent=[-2, 2, -2, 2], cmap="binary", interpolation="bilinear")
    plt.figure(figsize=(200, 200), dpi=250)
    plt.show()
```

```python
mandelBrot()
```

```python
def juliaSet():
    N = 300
    x = np.arange(-2, 2 + (3/N), 3/N) #This is the linspace of the x axis
    y = np.arange(-2, 2 + (3/N), 3/N)
    
    it = 50
    
    mag = list()
    
    #The julia set does not change it's cx or cy value
    for b in y:
        mag.append(list())
        for a in x:
            cy = 0.156
            cx = -0.8
            zx = a
            zy = b
            n = 0
            while n < it:
                nx = (zx**2) - (zy**2) + cx
                ny = (2 * zy * zx) + cy
                if (nx**2) + (ny**2) > 2:
                    mag[-1].append(0)
                    break
                zx = nx
                zy = ny
                n += 1
                
            if n == it:
                mag[-1].append(it)
                
    plt.imshow(mag, extent=[-2, 2, -2, 2], cmap="binary", interpolation="bilinear")
    plt.show()
```

```python
juliaSet()
```
